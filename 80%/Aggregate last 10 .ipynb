```python
#Desviación
df_saldo_no_rot_mdo = df.pivot_table('saldo_no_rot_mdo','id_cli', aggfunc={np.std})
df_saldo_no_rot_mdo.columns = ['saldo_no_rot_mdo']
df_saldo_no_rot_mdo.reset_index(inplace =True)
df_saldo_no_rot_mdo
```

```python
#Último valor
df_sorted = df.sort_values(by='periodo', ascending=False)
df_sorted.drop_duplicates(subset='id_cli', inplace = True)
df_pol_saldo_cupo = df_sorted[['id_cli','pol_centr_ext','saldo_prom3_tdc_mdo','cupo_tc_mdo','cant_oblig_tot_sf','cant_cast_ult_12m_sr']]
```

```python
df_sorted_2 = df.sort_values(by = ['id_cli','periodo'])
df_sorted_2
```

```python
df_sorted_subset =df_sorted_2[['id_cli','periodo','gasto_familiar']]
df_sorted_subset
```

```python
df_sorted_subset=df_sorted_subset.assign(count=df_sorted_subset.groupby(df_sorted_subset.id_cli.ne(df_sorted_subset.id_cli.shift()).cumsum()).cumcount().add(1))
#df_sorted_subset["Occur"] = df_sorted_subset["id_cli"].value_counts()
df_sorted_subset['ponderado'] = df_sorted_subset['gasto_familiar']*df_sorted_subset['count']
df_sorted_subset
```

```python
df_cuota_tc_mdo = df_sorted_subset.pivot_table(['count','ponderado'],'id_cli', aggfunc={np.sum})
df_cuota_tc_mdo.columns = ['count','ponderado']
df_cuota_tc_mdo.reset_index(inplace =True)
df_cuota_tc_mdo['gasto_familiar'] = df_cuota_tc_mdo['ponderado']/df_cuota_tc_mdo['count']
```

```python
df_gasto_familiar = df_cuota_tc_mdo[['id_cli','gasto_familiar']]
```

```python
df_sorted_subset_2 =df_sorted_2[['id_cli','periodo','ind']]
df_sorted_subset_2=df_sorted_subset_2.assign(count=df_sorted_subset_2.groupby(df_sorted_subset_2.id_cli.ne(df_sorted_subset_2.id_cli.shift()).cumsum()).cumcount().add(1))
#df_sorted_subset["Occur"] = df_sorted_subset["id_cli"].value_counts()
df_sorted_subset_2['ponderado'] = df_sorted_subset_2['ind']*df_sorted_subset_2['count']
df_sorted_subset_2
```

```python
df_cuota_ind = df_sorted_subset_2.pivot_table(['count','ponderado'],'id_cli', aggfunc={np.sum})
df_cuota_ind.columns = ['count','ponderado']
df_cuota_ind.reset_index(inplace =True)
df_cuota_ind['ind'] = df_cuota_ind['ponderado']/df_cuota_ind['count']
df_ind = df_cuota_ind[['id_cli','ind']]
```
