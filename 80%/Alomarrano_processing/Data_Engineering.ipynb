## Drop columns

```python
Todrop = ['departamento_residencia',
        'ciudad_laboral',
        'departamento_laboral',
        'tipo_vivienda',
        'rechazo_credito',
        'cat_ingreso',
        'cuota_cred_hipot',
        'convenio_lib',
        'ingreso_nomina',
        'ingreso_segurida_social',
         ]
```

```python
Totrain = Totrain.drop(columns=Todrop, axis=0)
Toevaluate = Toevaluate.drop(columns=Todrop, axis=0)
```

```python
To_replacena = ['mora_max',
                'cant_moras_30_ult_12_meses',
                'cant_moras_60_ult_12_meses',
                'cant_moras_90_ult_12_meses',
                'cant_mora_30_tdc_ult_3m_sf',
                'cant_mora_30_consum_ult_3m_sf',
                'cant_oblig_tot_sf',
                'cant_cast_ult_12m_sr',
                'ctas_activas',
                'ctas_embargadas']
```

```python
for i in To_replacena:
    Totrain[i] = Totrain[i].replace(np.nan,0)
    Toevaluate[i] = Toevaluate[i].replace(np.nan,0)
```

## Dropna

```python
Totrain = Totrain.dropna()
Totrain
```

## Concat

```python
Totrain['id_cli'] = Totrain['id_cli'].round().astype(str)
Totrain['periodo'] = Totrain['periodo'].round().astype(str)
Totrain['id_cli'] = [x[:-2] for x in Totrain['id_cli']]
Totrain['periodo'] = [x[:-2] for x in Totrain['periodo']]
```

```python
Totrain['id_registro'] = Totrain['id_cli']+'#'+Totrain['periodo']
Totrain
```
